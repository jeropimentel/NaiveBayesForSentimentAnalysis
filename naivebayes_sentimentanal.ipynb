```python
# transformar label en valor numerico
from sklearn.preprocessing import LabelEncoder 
label_encoding = LabelEncoder()


tips_df.iloc[:,[2,3,4,5]] = tips_df.iloc[:,[2,3,4,5]].apply(label_encoding.fit_transform)
```

```python
# usar one hot encoder after labelencoder para separar las categorias de cada col en tantas col como categorias hay
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer
```

```python
oh_encoding = ColumnTransformer([('OneHotEncoding', OneHotEncoder(drop='first'), [2,3,4,5])],remainder='passthrough')
tips_df_ohe = oh_encoding.fit_transform(tips_df)
tips_df_ohe
```

```python
# estandarizar el data frame del one hot encoder usando MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
tips_df_std = minmax.fit_transform(tips_df_ohe)
tips_df_std
```

```python
# standard scaling 
from sklearn.preprocessing import StandardScaler

zs = StandardScaler()
tips_df_std = zs.fit_transform(tips_df_ohe)
tips_df_std
```

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Jeronimo\Desktop\amazon_cells_labelled.txt", sep='\t', header=None)
data.head()                 
```

```python
# separar las dos col
X = data.iloc[:, 0]
y= data.iloc[:,1]
```

```python
# procesamiento del texto - primer paso usando CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
X_vec = vectorizer.fit_transform(X)
X_vec.todense()
```

```python
# ahora se pasa a tf-idf para darle un score a las palabras - una importancia
from sklearn.feature_extraction.text import TfidfTransformer

tfidf = TfidfTransformer()
X_tfidf = tfidf.fit_transform(X_vec)
X_tfidf = X_tfidf.todense()
X_tfidf
```

```python
# dividir en train and test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.25, random_state=0)
```

```python
# vamos a entrenar el modelo con naive bayes
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
# dato el fit, ahora podemos usar el y_predicte
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
clf.score(X_test, y_test) # mide la accuracy
```

```python
# sentiment analysis
from sklearn.svm import SVC # vector que separa los clusters

classifier = SVC(kernel='linear') 
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
classifier.score(X_test, y_test)
```

```python
# funcion para analisis de sentimientos
def sentiment_pred(classifier, training_matrix, doc):
    """function to predict the sentiment of a product review 
    classifier: pre trained model
    training_matrix: matrix of features associated with the trained model
    doc= product review whose sentiment needs to be identified"""
    X_new = training_matrix.transform(pd.Series(doc))
    X_new = X_new.todense()
    #do not use fit and transform because the model is fit
    from sklearn.feature_extraction.text import TfidfTransformer
    tfidf = TfidfTransformer()
    X_tfidf_new = tfidf.fit_transform(X_new)
    X_tfidf_new = X_tfidf_new.todense()
    
    y_new = classifier.predict(X_tfidf_new)
    if y_new[0] == 0:
        return "negative sentiment"
    elif y_new[0] == 1:
        return "positive sentiment"
    
```

```python
nb_clf = pickle.load(open("nb_sa", "rb")) # SVC linear entrenado con los review de amazon
vectorizer = pickle.load(open("vectorizer_sa", "rb")) # countvectorizer
new_doc = "The gadget works like a charm. Very satisfied with the product" 
sentiment_pred(nb_clf,vectorizer, new_doc)
```
